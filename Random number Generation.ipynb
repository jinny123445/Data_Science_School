```python
#가격이 10,000원인 주식이 있다. 
#이 주식의 일간 수익률(%)은 기댓값이 0%이고 
#표준편차가 1%인 표준 정규 분포를 따른다고 하자. 
#250일 동안의 주가를 무작위로 생성하라
revenue = np.random.randn(250)
revenue = 1+revenue/100
a = 10000
b = np.zeros_like(revenue)
for i in range(250):
    a *= revenue[i]
    b[i] = a
print(b)
```
